# Video Game Sales Exploratory Data Analysis

## Dataset Source
This analysis uses the cleaned dataset generated from the
Video Game Sales Data Cleaning project.


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_videogame_sales.csv')
df.head()
```

```python
df.shape
df.info()
df.describe()
```

## Key Questions
- Which platforms have the highest global sales?
- Which genres sell the most?
- How have global sales changed over the years?


## Sales by platforms

```python
platform_sales = df.groupby('platform')['global_sales'].sum().sort_values(ascending=False)
platform_sales = platform_sales.head(10)
platform_sales = platform_sales.sort_values(ascending=True)
plt.figure(figsize=(10,6))
plt.barh(platform_sales.head(10).index, platform_sales.head(10).values, color='#4e79a7')
plt.title('Top 10 Platforms by Global Sales', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Global Sales', fontsize=12)
plt.ylabel('Platform', fontsize=12)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(axis='x', linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

## Sales by Genre

```python
genre_sales = df.groupby('genre')['global_sales'].sum().sort_values(ascending=False)
genre_sales = genre_sales.head(10)
genre_sales = genre_sales.sort_values(ascending=True)
plt.figure(figsize=(10,6))
plt.barh(genre_sales.head(10).index, genre_sales.head(10).values, color='#4e79a7')
plt.title('Top 10 Genres by Global Sales', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Global Sales', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(axis='x', linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

## Sales Trend Over Years

```python
yearly_sales = df.groupby('year_of_release')['global_sales'].sum()
yearly_sales = yearly_sales.sort_index()
plt.figure(figsize=(10,5))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', color='#2c3e50',lw=2)
plt.title('Yearly Global Sales Trend', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Global Sales', fontsize=12)
plt.grid(True, linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

## Key Insights
- PlayStation 2 is the highest-selling platform globally.
- Action and Sports are the top-performing genres in terms of global sales.
- Global video game sales peaked between 2005 and 2010, corresponding to major console generations.
